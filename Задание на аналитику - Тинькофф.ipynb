```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

## Импорт данных:

```python
data = pd.read_excel('testovye-dannye-db81b75e-1bf9-4d69-b67d-f42fac4336f3.xlsx')

data = pd.DataFrame(data, columns=['campaign_nm','stat_date', 'shows', 'clicks', 'spent','заявки','одобрения','оформление продукта'])

data
```

## Сгруппируем данные по названию кампании:

```python
data = data.groupby('campaign_nm').sum()
data
```

## Посчитаем конверсию и CPO:

```python
data['Конверсия (%)'] = (data['оформление продукта'] / data['shows'])*100
data['CPO'] = data['spent'] / data['оформление продукта']
data 
```

## Выведем ТОП-5 кампаний по возрастанию CPO:

```python
data.sort_values('CPO')[:5]
```
